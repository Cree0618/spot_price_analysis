```python
modified_df = pd.read_csv("src\MAIN_DATAFRAME.csv", sep=",")
modified_df
```

```python
data = modified_df
```

```python
# Calculate the average current price difference between SPOT and FORWARD prices
average_price_difference = data['Price_difference'].mean()

# Create a new column for increased SPOT prices (e.g., 10% increase)
data['Increased_SPOT_Price'] = data['Cost on SPOT'] * 1.10

# Calculate the new price difference with increased SPOT prices
data['New_Price_difference'] = data['Increased_SPOT_Price'] - data['Cost on FORWARD']

# Display the results
average_price_difference, data[['Cost on SPOT', 'Cost on FORWARD', 'Increased_SPOT_Price', 'New_Price_difference']].head()
```

```python
import matplotlib.pyplot as plt

# Calculate the overall statistics for the entire period
overall_average_spot = data['Cost on SPOT'].mean()
overall_average_forward = data['Cost on FORWARD'].mean()
overall_average_increased_spot = data['Increased_SPOT_Price'].mean()
overall_average_new_price_diff = data['New_Price_difference'].mean()

# Plot the prices and differences over the entire period
plt.figure(figsize=(12, 6))
plt.plot(data['Interval_Start'], data['Cost on SPOT'], label='SPOT Price', alpha=0.7)
plt.plot(data['Interval_Start'], data['Cost on FORWARD'], label='FORWARD Price', alpha=0.7)
plt.plot(data['Interval_Start'], data['Increased_SPOT_Price'], label='Increased SPOT Price (10%)', alpha=0.7)
plt.xlabel('Interval Start')
plt.ylabel('Price (Kč)')
plt.title('SPOT vs FORWARD Prices Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(data['Interval_Start'], data['Price_difference'], label='Original Price Difference', alpha=0.7)
plt.plot(data['Interval_Start'], data['New_Price_difference'], label='New Price Difference (10% SPOT Increase)', alpha=0.7)
plt.xlabel('Interval Start')
plt.ylabel('Price Difference (Kč)')
plt.title('Price Difference Between SPOT and FORWARD Prices Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display overall statistics
overall_stats = {
    'Overall Average SPOT Price': overall_average_spot,
    'Overall Average FORWARD Price': overall_average_forward,
    'Overall Average Increased SPOT Price': overall_average_increased_spot,
    'Overall Average New Price Difference': overall_average_new_price_diff
}
```

```python
overall_stats
```

```python
# Take modified_df and add row that compares the price of SPOT and FORWARD and says what is cheaper

# If SPOT is cheaper, add 1 to the new column "Cheaper"
# If FORWARD is cheaper, add 0 to the new column "Cheaper"

modified_df['Cheaper'] = np.where(modified_df['Cost on SPOT'] < modified_df['Cost on FORWARD'], 1, 0)
```

```python
# Plotting the Price_difference data with conditional coloring
import matplotlib.pyplot as plt

# Extract the Price_difference column
price_difference = data['Price_difference']

# Plotting the Price_difference data
plt.figure(figsize=(10, 6))
plt.plot(price_difference, label='Price Difference')
plt.xlabel('Index')
plt.ylabel('Price Difference')
plt.title('Price Difference Over Time')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(price_difference, label='Price Difference', color='blue')
plt.plot(price_difference[price_difference < 0], color='red', label='Negative Price Difference')
plt.xlabel('Index')
plt.ylabel('Price Difference')
plt.title('Price Difference Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

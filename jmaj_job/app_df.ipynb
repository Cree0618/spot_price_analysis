```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Load the new data
file_path = 'originalmesicni.csv'
data = pd.read_csv(file_path)

# Ensure data integrity by checking for duplicates and only working with the loaded rows
data = data.drop_duplicates()

```

```python
# remove rows 1-6 from "data" and save it to "data2"
data_modified = data.drop(data.index[0:8])
```

```python
# Convert 'Date' column to datetime if necessary
# Assuming the date column is named 'Date' and is in 'YYYY-MM' format

data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')

# Extract year and month from the 'Date' column
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

# Define a function to determine the quarter
def get_quarter(month):
    if month in [1, 2, 3]:
        return 'Q1'
    elif month in [4, 5, 6]:
        return 'Q2'
    elif month in [7, 8, 9]:
        return 'Q3'
    elif month in [10, 11, 12]:
        return 'Q4'

# Apply the function to create a 'Quarter' column
data['Quarter'] = data['Month'].apply(get_quarter)
```
